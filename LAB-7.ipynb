```python
#task13
def calic(opr,ope1,ope2):
  if(opr=="+"):
    res=ope1+ope2
  elif(opr=="-"):
    res=ope1-ope2
  elif(opr=="*"):
    res=ope1*ope2
  elif(opr=="/"):
    res=ope1/ope2
  return res
opr=input()
ope1=float(input())
ope2=float(input())
calic(opr,ope1,ope2)
```

```python
#task14
def karma(X,n):
  p=""
  r=""
  for i in range(0,len(X),1):
    if(not(i==0)):
      if(i%n==0):
        r=r+X[i]
        p=p
      else:
        p=p+X[i]
        r=r
  return (X[0]+p+r)
inp=input()
num=int(input())
karma(inp,num)
```

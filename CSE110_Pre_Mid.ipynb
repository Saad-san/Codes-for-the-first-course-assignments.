Your task is to create a python program that takes as input an integer. Then take as input the marks of that many students and calculate the sum of all the students who scored more than 50, and calculate the average of all the students who scored more than 30. Finally show the sum and average.


Sample Input 2:
6

30

51

50

29

31

99

Sample Output 2:
150
57.75

Explanation 2:
The first input is 6, So, the next inputs are the marks of 6 students. The sum of the marks greater than 50 is 150 (51+99) (the first output). The average of the marks greater than 30 is 57.75 ((51+50+31+99)/4), which is the second output.






````python
n=int(input())
i=0
sum=0
a=0
b=0
while(i<n):
  num=int(input())
  if(num>50):
    sum+=num
  if(num>30):
    a+=num
    b+=1
  i=i+1
print(sum)
print(a/b)
````

#Checking Password
Write a Python program that takes a string input and check if it is a valid password or not.

Condition for password:
1.Atleast 2 uppercase letter

2.Atlest 2 lowercase letter

3.Atleast 3 digit

4.Atleast one special character(#@!)


Sample Input1:abC1@q214DL

Sample Output: Valid

````python
s=input()
uc=0
sc=0
num=0
sc=0
digit="0123456789"
special="@#!"
char=0
for i in s:
  if(ord('a')<=ord(i)<=ord('z')):           #sc
    sc+=1
  if(ord('A')<=ord(i)<=ord('Z')):
    uc+=1
  if (i in digit):
    num+=1
  if(i in special):
    char+=1
if(uc>=2 and sc>=2 and num>=3 and char>=1):
  print("valid")
else:
  print("invalid")
````
